```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
```

```python
browser = webdriver.Firefox()
```

```python
browser.get(url)
```

```python
html = browser.page_source
html
```

```python
soup = BeautifulSoup(html,'html.parser')
soup
```

```python
type(soup)
```

```python
soup.find_all('a',href="https://instagram.com/utm_medium=copy_link")
```

```python
soup.select('figure.wp-block-table.is-style-stripes')[1].find_all('thead')[0].find_all('th')[0].text.strip()
```

```python
table = soup.select('figure.wp-block-table.is-style-stripes')[1]
```

```python
thead = [i.text.strip() for i in table.find_all('th')]
thead
```

```python
import pandas as pd
```

```python
def shark():
    thead = [i.text.strip() for i in table.find_all('th')]
    data=[]
    for a in soup.select('figure.wp-block-table.is-style-stripes')[1].tbody.find_all('tr'):
        data.append([i.text.strip() for i in a.find_all('td')])
    df = pd.DataFrame(data)
    df.columns = thead
    return df
```

```python
df1 = shark()
df1.head()
```

```python
df1.info()
```

```python
df1["Sr No."] = pd.to_numeric(df1["Sr No."])
```

```python
df1.info()
```

```python
df1.to_csv("Shark_Tank_Investor_Details.csv",index=False)
```

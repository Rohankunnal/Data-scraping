```python
import pandas as pd
def flipkart():
    prod_name = [i.text.strip() for i in soup.select('div._4rR01T')]
    price = [''.join((i.text.strip().split('₹')[1].split(','))) for i in soup.select('div._25b18c')]
    rate = [i.find_all('div')[0].text.strip() for i in soup.select('div.gUuXy-')]
    total_rate = [''.join(i.text.strip().split(' ')[0].split(',')) for i in soup.select('span._2_R_DZ')]
    total_rev = [i.text.split('Ratings\xa0&\xa0')[1].split(' ')[0] for i in soup.select('span._2_R_DZ')]
    spec = [i.text for i in soup.select('div.fMghEO')]
    return pd.DataFrame({'Product Name':prod_name,
                         'Price':price,
                         'Rating out of 5':rate,
                         'Total Ratings':total_rate,
                         'Total Reviews':total_rev,
                         'Device Features':spec})
```

```python
tables=[]
from tqdm import tqdm
for i in range(1,3):
    urls=url.format(i)
    browser.get(url)
    html=browser.page_source
    soup=BeautifulSoup(html,'lxml')
    tables.append(flipkart())
```

```python
df=pd.concat(tables,axis=0)
df.head()
```
